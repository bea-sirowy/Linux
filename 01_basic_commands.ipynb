Finding yourself with pwd

Checking your user name with whoami

Changing directories with cd

cd /[the name of the directory]

To move up one level we use cd .. 

To move up more than one level, use multiple double-dot pairs

cd .. .. 

cd .. .. ..

To move up to the root level in the file structure entering cd /


To see the content of a directory use ls (similar to dir command in windows)

To get more info about the directories use ls -l (long listing)

To view the hidden files type ls -la

To access help on a given tool - e.g. aircrack-ng - type  --help

Other options to access help: -h  or -?

Most commands have a manual (man) page with more info. To access it type man

e.g. man python

To exit: type q

To find an instance of a given string type locate
e.g. locate aircrack-ng

If you are looking for a binary file you can use whereis

eg. whereis aircrack-ng

To find the path of a file type which
e.g. which aircrack-ng

Performing more powerful searches with find:

find _directory options expressions_
